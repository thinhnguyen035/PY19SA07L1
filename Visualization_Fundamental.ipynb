```python
# quantity by week_num, branch
sale = pd.read_excel('supermarket_sales_nan.xlsx')
sale['Date'] = pd.to_datetime(sale['Date'])
sale['week_num'] = sale['Date'].dt.isocalendar().week

# sale

# sale['Date'].dt.year

sale_by_period = sale.groupby(['week_num','Branch'],as_index=False)['Quantity'].sum()
sale_by_period
sale_m1 = sale_by_period[sale_by_period['week_num']<=4]
sale_m1
```

## Bước 5: Add labels, text

```python
# def addlabels(xs, ys):
#     for x, y in zip(xs, ys):
#         label = "{:.0f}".format(y)  # Định dạng số y thành chuỗi, không có số thập phân
#         plt.annotate(
#             label,           # Nội dung hiển thị
#             xy=(x, y),       # Tọa độ điểm cần gán nhãn
#             xytext=(x, y-2), # Dời nhãn lên/xuống (ở đây là xuống 2 đơn vị)
#             ha='center'      # Căn giữa nhãn theo chiều ngang
#         )

def addlabels(xs,ys):
    for x,y in zip(xs,ys):

        label = "{:.0f}".format(y)

        plt.annotate(label, # this is the text
                     xy=(x,y), # these are the coordinates to position the label
                     xytext=(x,y-2),
                     ha='center')
```

```python
# Hiển thị một đoạn văn bản (text) tại vị trí (x, y + 5), tức là lệch lên trên một chút so với điểm dữ liệu thực sự (x, y).

# Kèm theo mũi tên chỉ từ đoạn văn bản đến đúng điểm (x, y) trên biểu đồ.
# ef addtext(x, y, text):
#     plt.annotate(
#         text,              # nội dung văn bản cần hiển thị
#         xy=(x, y),         # tọa độ của điểm cần chú thích
#         xytext=(x, y+5),   # vị trí hiển thị đoạn text (cách điểm đó 5 đơn vị theo trục y)
#         color='red',       # màu chữ là đỏ
#         arrowprops=dict(   # cấu hình mũi tên
#             facecolor='black',    # màu mũi tên là đen
#             headwidth=4,          # chiều rộng đầu mũi tên
#             width=2,              # độ dày thân mũi tên
#             headlength=4          # độ dài đầu mũi tên
#         ),
#         horizontalalignment='left',  # căn lề chữ theo chiều ngang
#         verticalalignment='top'      # căn lề chữ theo chiều dọc
#     )



def addtext(x,y,text):
    plt.annotate(text, # this is the text
                 xy=(x,y), # these are the coordinates to position the label
                 xytext=(x,y+5),
                 color='red',
                 arrowprops=dict(facecolor='black',headwidth=4,width=2,headlength=4),
                 horizontalalignment='left',verticalalignment='top')
```

```python
# Add text & label

sale_m1_4w = sale_m1[sale_m1['week_num']<5]
x = sale_m1_4w['week_num'].unique().astype('str')
A = sale_m1_4w[sale_m1_4w['Branch']=='A']['Quantity'].unique()
B = sale_m1_4w[sale_m1_4w['Branch']=='B']['Quantity'].unique()
C = sale_m1_4w[sale_m1_4w['Branch']=='C']['Quantity'].unique()

values = ['Week_'+ n for n in x]

plt.plot(x,A,label='Aoen')
plt.plot(x,B,label='BigC')
plt.plot(x,C,label='Coop')

plt.title('Quantity by Week and Super Market', fontsize=10)
plt.xlabel('Week Number', fontsize=10)
plt.ylabel('Quantity', fontsize=10)

plt.legend(loc="upper left")
plt.xticks(x,values)

addlabels(x,A)
addlabels(x,B)
addlabels(x,C)

addtext(2,189,'Peak of Aoen')
addtext(3,175,'Peak of BigC')
addtext(3,144,'Peak of Coop')

plt.show()
```
